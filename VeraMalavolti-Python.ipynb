```python
from datetime import datetime
from collections import Counter
import pprint

# Dati esterni: marche e taglie
MARCHE_VESTITI = [
    "Abercrombie & Fitch", "Adidas", "Alexander Mcqueen", "Amisu", "Armani", "Asos", "Balenciaga", "Balmain", "Barbour", "Benetton",
    "Bershka", "Bottega Veneta", "Brunello Cucinelli", "Burberry", "C&A", "Cache Cache", "Calliope", "Calvin Klein", "Canada Goose",
    "Carhartt", "Celine", "Champion", "Chanel", "Columbia", "Converse", "Costes", "Desigual", "Diesel", "Dolce & Gabbana", "Dr. Martens",
    "Dsquared2", "Ellesse", "Ermenegildo Zegna", "Esprit", "Etam", "Etro", "Fb Sister", "Fendi", "Fila", "Fred Perry", "Gant", "Geox",
    "Givenchy", "Gucci", "Guess", "H&M", "H&M Divided", "Herm√®s", "Hollister", "Hugo Boss", "Jack & Jones", "Jennyfer", "Kappa", "Kenzo",
    "Kiabi", "Lacoste", "Lefties", "Levi's", "Loavies", "Loewe", "Louis Vuitton", "Made In Italy", "Mango", "Marc Jacobs", "Massimo Dutti",
    "Max Mara", "Miu Miu", "Moncler", "Morgan", "Moschino", "Motivi", "Naf Naf", "Napapijri", "New Balance", "New Collection",
    "New Yorker", "Nike", "North Face", "Off-White", "Only", "Orignal Marines", "Ovs", "Patagonia", "Peuterey", "Piazza Italia", "Pimkie",
    "Pinko", "Prada", "Primark", "Promod", "Pull&Bear", "Puma", "Ralph Lauren", "Reebok", "Saint Laurent", "Salvatore Ferragamo",
    "Sandro", "Scotch & Soda", "Selected", "Set", "Shein", "Skechers", "Stone Island", "Stradivarius", "Subdued", "Superdry", "Supreme",
    "Tally Weijl", "Terranova", "Tezenis", "Timberland", "Tod's", "Tom Ford", "Tommy Hilfiger", "Under Armour", "Uniqlo",
    "United Colors Of Benetton", "Valentino", "Vans", "Vero Moda", "Versace", "Vintage Dressing", "Zara", "Zuiki"
]

COLORI_VALIDI = [
    "Nero", "Blu", "Blu marino", "Grigio", "Bianco", "Beige", "Marrone", "Azzurro", 
    "Cachi", "Rosso", "Panna", "Rosa", "Multicolore", "Verde", "Verde scuro", 
    "Borgogna", "Giallo", "Arancione", "Viola", "Senape", "Turchese", "Corallo", 
    "Lilla", "Menta", "Oro", "Argento", "Albicocca"
]

def valida_colore(colore):
    if colore.capitalize() not in COLORI_VALIDI:
        print("Spiacente! Scegli tra uno dei nostri colori: Nero, Blu, Blu marino, Grigio, Bianco, Beige, Marrone, Azzurro, Cachi, Rosso, Panna, Rosa, Multicolore, Verde, Verde scuro, Borgogna, Giallo, Arancione, Viola, Senape, Turchese, Corallo, Lilla, Menta, Oro, Argento, Albicocca.")
        return False
    return True
    
STAGIONI_VALIDE = ["Primavera", "Estate", "Autunno", "Inverno"]

def valida_stagione(stagione):
    if stagione.capitalize() not in STAGIONI_VALIDE:
        print("Ops! Scegli una delle nostre stagioni: Primavera, Estate, Autunno, Inverno.")
        return False
    return True


CATEGORIE_TIPO_VALIDE = [
    "Poncho", "Cappotto", "Gilet", "Giacca", "Felpa", "Maglione", "Kimono", "Cardigan", "Bolero", "Panciotto",
    "Abito", "Tubino", "Denim", "Gonna", "T-shirt", "Canotta", "Tunica", "Top", "Body", "Dolcevita",
    "Jeans", "Pantalone", "Leggings", "Pantaloncini", "Tuta", "Jumpsuit", "Costume da bagno", "Bikini",
    "Copricostume", "Lingerie", "Pigiama", "Collant", "Calzini", "Vestaglia", "Camicia", "Canottiera", "Pullover", "Sciarpa", "Foulard"
]

def valida_tipo(tipo):
    if tipo.capitalize() not in CATEGORIE_TIPO_VALIDE:
        print(f"Spiacente! Indica una di queste categorie per continuare a registrare il tuo prodotto: {', '.join(CATEGORIE_TIPO_VALIDE)}.")
        return False
    return True


TAGLIE_VALIDE = [
    "XXXS / EU 30",
    "XXS / EU 32",
    "XS /  EU 34",
    "S / EU 36",
    "M / EU 38",
    "L / EU 40",
    "XL / EU 42",
    "XXL / EU 44",
    "XXXL / EU 46",
    "4XL / EU 48",
    "5XL / EU 50",
    "6XL / EU 52",
    "7XL / EU 54",
    "8XL / EU 56",
    "9XL / EU 58",
    "Taglia unica"
]

def valida_taglia(input_taglia):
    input_taglia = input_taglia.strip().upper()
    for taglia in TAGLIE_VALIDE:
        if input_taglia in taglia.upper():
            return taglia  # ritorna la versione completa
    print("Ops! Prova con una delle nostre taglie valide: XXXS / EU 30, XXS / EU 32, XS / EU 34, S / EU 36, M / EU 38, L / EU 40, XL / EU 42, XXL / EU 44, XXXL / EU 46, 4XL / EU 48, 5XL / EU 50, 6XL / EU 52, 7XL / EU 54, 8XL / EU 56, 9XL / EU 58, Taglia unica.")
    return None

# Funzioni utili
def is_marca_valida(marca):
    return marca in MARCHE_VESTITI

def suggerisci_marche(parziale):
    return [m for m in MARCHE_VESTITI if m.lower().startswith(parziale.lower())]

def accetta_parola(parola_corretta, parola_inserita, soglia_errore=0.7):
    """
    Funzione per accettare una parola con una certa tolleranza sugli errori.
    In questo caso, consideriamo le parole che cominciano con lo stesso prefisso.
    """
    return parola_corretta.lower().startswith(parola_inserita.lower())

# Collezione principale
collezione = []

MATERIALI_VALIDI = [
    "Alpaca", "Bamb√π", "Camoscio", "Carta", "Cashmere", "Chiffon", "Cotone", "Denim", "Elastane", "Feltro", "Finta pelliccia", "Flanella", "Juta", "Lana", "Lino", "Merino", "Mohair", "Neoprene", "Nylon", "Paillette", "Pelle", "Pelle verniciata", "Pile", "Piumino", "Pizzo", "Plastica", "Poliestere", "Raso", "Rete", "Seta", "Tulle", "Tweed", "Velluto", "Velluto a coste", "Velour", "Viscosa"
]

def valida_materiale(materiale):
    if materiale.capitalize() not in MATERIALI_VALIDI:
        print("Ops! Materiale non valido. Scegli tra uno dei nostri materiali: Alpaca, Bamb√π, Camoscio, Carta, Cashmere, Chiffon, Cotone, Denim, Elastane, Feltro, Finta pelliccia, Flanella, Juta, Lana, Lino, Merino, Mohair, Neoprene, Nylon, Paillette, Pelle, Pelle verniciata, Pile, Piumino, Pizzo, Plastica, Poliestere, Raso, Rete, Seta, Tulle, Tweed, Velluto, Velluto a coste, Velour, Viscosa.")
        return False
    return True

def valida_marca(marca):
    if marca.capitalize() not in [m.capitalize() for m in MARCHE_VESTITI]:
        print("Ops! Scegli una delle nostre marche: Abercrombie & Fitch, Adidas, Alexander Mcqueen, Amisu, Armani, Asos, Balenciaga, Balmain, Barbour, Benetton, Bershka, Bottega Veneta, Brunello Cucinelli, Burberry, C&A, Cache Cache, Calliope, Calvin Klein, Canada Goose, Carhartt, Celine, Champion, Chanel, Columbia, Converse, Costes, Desigual, Diesel, Dolce & Gabbana, Dr. Martens, Dsquared2, Ellesse, Ermenegildo Zegna, Esprit, Etam, Etro, Fb Sister, Fendi, Fila, Fred Perry, Gant, Geox, Givenchy, Gucci, Guess, H&M, H&M Divided, Herm√®s, Hollister, Hugo Boss, Jack & Jones, Jennyfer, Kappa, Kenzo, Kiabi, Lacoste, Lefties, Levi's, Loavies, Loewe, Louis Vuitton, Made In Italy, Mango, Marc Jacobs, Massimo Dutti, Max Mara, Miu Miu, Moncler, Morgan, Moschino, Motivi, Naf Naf, Napapijri, New Balance, New Collection, New Yorker, Nike, North Face, Off-White, Only, Orignal Marines, Ovs, Patagonia, Peuterey, Piazza Italia, Pimkie, Pinko, Prada, Primark, Promod, Pull&Bear, Puma, Ralph Lauren, Reebok, Saint Laurent, Salvatore Ferragamo, Sandro, Scotch & Soda, Selected, Set, Shein, Skechers, Stone Island, Stradivarius, Subdued, Superdry, Supreme, Tally Weijl, Terranova, Tezenis, Timberland, Tod's, Tom Ford, Tommy Hilfiger, Under Armour, Uniqlo, United Colors Of Benetton, Valentino, Vans, Vero Moda, Versace, Vintage Dressing, Zara, Zuiki.")
        return False
    return True

def registra_elemento(tipo, colore, taglia, materiale, stagione, marca, prezzo, immagine):
    # taglia = taglia.upper()  # Non serve pi√π, la taglia √® gi√† validata e formattata
    if taglia not in TAGLIE_VALIDE:
        raise ValueError("Taglia non valida.")

    if not valida_marca(marca):
        raise ValueError("Marca non valida.")

    if not valida_materiale(materiale):
        raise ValueError("Materiale non valido.")

    elemento = {
        "tipo": tipo,
        "colore": colore,
        "taglia": taglia,
        "materiale": materiale,
        "stagione": stagione,
        "marca": marca,
        "prezzo": prezzo,
        "immagine": immagine,
        "timestamp": datetime.now()
    }

    collezione.append(elemento)
    print("‚úÖ Elemento registrato correttamente!")


def visualizza_elementi():
    print("\nüëï TUTTI I CAPI INSERITI (dal pi√π recente al pi√π vecchio):")
    ordinati = sorted(collezione, key=lambda x: x["timestamp"], reverse=True)
    for idx, e in enumerate(ordinati, 1):
        print(f"\nüîπ CAPI N. {idx}")
        pprint.pprint(e)

def cerca_elementi(campo, valore):
    if campo.lower() == "colore":
        if not valida_colore(valore):
            return
    if campo.lower() == "materiale":
        if not valida_materiale(valore):
            return
    if campo.lower() == "marca":
        if not valida_marca(valore):
            return
    if campo.lower() == "tipo":
        if not valida_tipo(valore):
            return
    if campo.lower() == "taglia":
        if not valida_taglia(valore):
            return
    valore = valore.capitalize()
    risultati = [e for e in collezione if str(e.get(campo, "")).lower() == valore.lower()]
    print(f"\nüîç Trovati {len(risultati)} risultati per {campo} = {valore}")
    for r in risultati:
        pprint.pprint(r)

def filtra_elementi(tipo="", colore="", taglia="", materiale="", marca="", stagione="", prezzo_massimo=None):
    tipo = tipo.capitalize()
    colore = colore.capitalize()
    materiale = materiale.capitalize()
    marca = marca.capitalize()
    stagione = stagione.capitalize()
    taglia = taglia.upper()

    if tipo and not valida_tipo(tipo):
        return
    if colore and not valida_colore(colore):
        return
    if materiale and not valida_materiale(materiale):
        return
    if marca and not valida_marca(marca):
        return
    if taglia and not valida_taglia(taglia):
        return

    risultati = collezione
    if tipo:
        risultati = [e for e in risultati if e['tipo'] == tipo]
    if colore:
        risultati = [e for e in risultati if e['colore'] == colore]
    if materiale:
        risultati = [e for e in risultati if e['materiale'] == materiale]
    if marca:
        risultati = [e for e in risultati if e['marca'] == marca]
    if taglia:
        risultati = [e for e in risultati if e['taglia'] == taglia]
    if stagione:
        risultati = [e for e in risultati if e['stagione'] == stagione]
    if prezzo_massimo is not None:
        risultati = [e for e in risultati if e['prezzo'] <= prezzo_massimo]

    print(f"\nüîé Trovati {len(risultati)} capi che soddisfano i criteri.")
    for r in risultati:
        pprint.pprint(r)

def statistiche():
    print("\nüìä STATISTICHE SUI CAPI INSERITI")
    print(f"Totale capi inseriti: {len(collezione)}")
    if not collezione:
        return

    tipi = Counter([e["tipo"] for e in collezione])
    colori = Counter([e["colore"] for e in collezione])
    taglie = Counter([e["taglia"] for e in collezione])
    stagioni = Counter([e["stagione"] for e in collezione])
    prezzi = [e["prezzo"] for e in collezione]

    print("Distribuzione per tipo:", dict(tipi))
    print("Distribuzione per colore:", dict(colori))
    print("Distribuzione per taglia:", dict(taglie))
    print("Distribuzione per stagione:", dict(stagioni))
    print(f"Prezzo medio: {sum(prezzi)/len(prezzi):.2f} ‚Ç¨")

    pi√π_recente = max(collezione, key=lambda x: x["timestamp"])
    print("üëï Capo pi√π recente aggiunto:")
    pprint.pprint(pi√π_recente)

PARAMETRI_RICERCA_SEMPLICE = [
    "Tipo", "Colore", "Taglia", "Materiale", "Stagione", "Marca", "Prezzo", "Immagine"
]
def menu():
    while True:
        print("\nüì≤ MENU LOOKBOOK")
        print("1. Registra nuovo elemento")
        print("2. Visualizza tutti gli elementi")
        print("3. Ricerca semplice")
        print("4. Filtraggio avanzato")
        print("5. Statistiche")
        print("6. Esci")
        print("-" * 40)

        scelta = input("Scegli un'opzione: ").strip()

#Potrebbe succedere in questa sezione che l'utente inserisca un testo non compreso tra quelli previsti dal codice. In questo caso compare una stringa di testo che indica quali sono le opzioni corrette che pu√≤ inserire per procedere la sua attivit√†
#Ho previsto che le opzioni possano anche essere lasciate vuote se l'utente non ha un'informazione tra quelle inseribili per la registrazione (es. la taglia)
        if scelta == "1":
            while True:
                tipo = input("Tipo di capo (es. T-shirt): ").strip()
                if valida_tipo(tipo):
                    break
            while True:
                colore = input("Colore: ").strip()
                if valida_colore(colore):
                    break
            while True:
                input_taglia = input("Taglia (es. 38, M, EU 38): ").strip()
                taglia_validata = valida_taglia(input_taglia)
                if taglia_validata:
                    taglia = taglia_validata
                    break
            while True:
                materiale = input("Materiale: ").strip()
                if valida_materiale(materiale):
                    break
            while True:
                stagione = input("Stagione (es. Estate): ").strip()
                if valida_stagione(stagione):
                    break
            while True:
                marca = input("Marca: ").strip()
                if valida_marca(marca):
                    break
            prezzo = float(input("Prezzo: ").strip())
            immagine = input("URL immagine: ").strip()
            registra_elemento(tipo, colore, taglia, materiale, stagione, marca, prezzo, immagine)
            
        elif scelta == "2":
            visualizza_elementi()

#Potrebbe succedere in questa sezione che l'utente inserisca un testo non compreso tra quelli previsti dal codice. In questo caso compare una stringa di testo che indica quali sono le opzioni corrette che pu√≤ inserire per procedere la sua attivit√†
#Ho previsto che le opzioni possano anche essere lasciate vuote se l'utente non sta cercando un'informazione tra quelle possibili
        elif scelta == "3":
            while True:
                print(f"Parametri disponibili: {', '.join(PARAMETRI_RICERCA_SEMPLICE)}")
                campo = input("Campo di ricerca (es. tipo, colore, taglia): ").strip()
                if campo.capitalize() not in PARAMETRI_RICERCA_SEMPLICE:
                    print("Opzione non valida! Scegli tra uno dei parametri di ricerca suggeriti: Tipo, Colore, Taglia, Materiale, Stagione, Marca, Prezzo, Immagine")
                    continue
                while True:
                    valore = input(f"Valore da cercare per '{campo}': ").strip()
                    # Validazione interattiva
                    if campo.lower() == "tipo" and not valida_tipo(valore):
                        continue
                    if campo.lower() == "colore" and not valida_colore(valore):
                        continue
                    if campo.lower() == "taglia" and not valida_taglia(valore):
                        continue
                    if campo.lower() == "materiale" and not valida_materiale(valore):
                        continue
                    if campo.lower() == "marca" and not valida_marca(valore):
                        continue
                    if campo.lower() == "stagione" and not valida_stagione(valore):
                        continue
                    break
                cerca_elementi(campo, valore)
                break

#Potrebbe succedere in questa sezione che l'utente inserisca un testo non compreso tra quelli previsti dal codice. In questo caso compare una stringa di testo che indica quali sono le opzioni corrette che pu√≤ inserire per procedere la sua attivit√†
#Ho previsto che le opzioni possano anche essere lasciate vuote se l'utente non sta cercando un'informazione tra quelle possibili

        elif scelta == "4":
            while True:
                tipo = input("Tipo di capo (opzionale): ").strip()
                if tipo and not valida_tipo(tipo):
                    continue
                break
            while True:
                colore = input("Colore (opzionale): ").strip()
                if colore and not valida_colore(colore):
                    continue
                break
            while True:
                taglia = input("Taglia (opzionale): ").strip()
                if taglia and not valida_taglia(taglia):
                    continue
                break
            materiale = input("Materiale (opzionale): ").strip()
            while True:
                marca = input("Marca (opzionale): ").strip()
                if marca and not valida_marca(marca):
                    continue
                break
            while True:
                stagione = input("Stagione (opzionale): ").strip()
                if stagione and not valida_stagione(stagione):
                    continue
                break
            prezzo_massimo = input("Prezzo massimo (opzionale): ").strip()
            prezzo_massimo = float(prezzo_massimo) if prezzo_massimo else None
            filtra_elementi(tipo, colore, taglia, materiale, marca, stagione, prezzo_massimo)

        elif scelta == "5":
            statistiche()

        elif scelta == "6":
            print("üëã Grazie per aver utilizzato il Lookbook!")
            break

        else:
            print("‚ùå Opzione non valida. Riprova.")

if __name__ == "__main__":
        menu()
```
